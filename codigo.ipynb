Nesse trabalho irei implementar o método de ordenação Quick Sort e Insertion Sort.

```python
import random
import time
import matplotlib.pyplot as plt


def particao(vetor, baixo, alto):
    i = baixo - 1
    pivo = vetor[alto]

    for j in range(baixo, alto):
        if vetor[j] <= pivo:
            i += 1
            vetor[i], vetor[j] = vetor[j], vetor[i]

    vetor[i + 1], vetor[alto] = vetor[alto], vetor[i + 1]
    return i + 1


def quicksort(vetor, baixo, alto):
    while baixo < alto:
        indice_pivo = particao(vetor, baixo, alto)

        # Ordena a menor parte primeiro para evitar chamadas recursivas excessivas
        if indice_pivo - baixo < alto - indice_pivo:
            quicksort(vetor, baixo, indice_pivo - 1)
            baixo = indice_pivo + 1
        else:
            quicksort(vetor, indice_pivo + 1, alto)
            alto = indice_pivo - 1


def insertionsort(vetor):
    tamanho = len(vetor)

    for i in range(1, tamanho):
        chave = vetor[i]
        j = i - 1

        while j >= 0 and vetor[j] > chave:
            vetor[j + 1] = vetor[j]
            j -= 1

        vetor[j + 1] = chave
```

```python


tamanhos = [50, 500, 5000, 50000] 
TempoQuickSortCres = []
TempoQuickSortDecres = []
TempoQuickSortAlea = []
TempoInsertionSortCres = []
TempoInsertionSortDecres = []
TempoInsertionSortAlea = []

for tam in tamanhos:
    VetCrescenteQS = list(range(tam))
    VetCrescenteIS = list(range(tam))

    inicio = time.time()
    quicksort(VetCrescenteQS, 0, len(VetCrescenteQS) - 1)
    fim = time.time()
    tempo_execucao = fim - inicio
    TempoQuickSortCres.append(tempo_execucao)

    inicio = time.time()
    insertionsort(VetCrescenteIS)
    fim = time.time()
    tempo_execucao = fim - inicio
    TempoInsertionSortCres.append(tempo_execucao)


    VetDecrescenteQS = list(range(tam, 0, -1))
    VetDecrescenteIS = list(range(tam, 0, -1))

    inicio = time.time()
    quicksort(VetDecrescenteQS, 0, len(VetDecrescenteQS) - 1)
    fim = time.time()
    tempo_execucao = fim - inicio
    TempoQuickSortDecres.append(tempo_execucao)

    inicio = time.time()
    insertionsort(VetDecrescenteIS)
    fim = time.time()
    tempo_execucao = fim - inicio
    TempoInsertionSortDecres.append(tempo_execucao)
    
    VetAleatorioQS = random.sample(range(tam), tam)
    VetAleatorioIS = random.sample(range(tam), tam)

    inicio = time.time()
    quicksort(VetAleatorioQS, 0, len(VetAleatorioQS) - 1)
    fim = time.time()
    tempo_execucao = fim - inicio
    TempoQuickSortAlea.append(tempo_execucao)

    inicio = time.time()
    insertionsort(VetAleatorioIS)
    fim = time.time()
    tempo_execucao = fim - inicio
    TempoInsertionSortAlea.append(tempo_execucao)

plt.plot(tamanhos, TempoQuickSortCres, label='Quick Sort', marker='.')
plt.plot(tamanhos, TempoInsertionSortCres, label='Insertion Sort', marker='.')
plt.ylabel('Tempo de execução (s)')
plt.xlabel('Tamanho do vetor')
plt.title('Vetor Crescente')
plt.legend()
plt.show()

plt.plot(tamanhos, TempoQuickSortDecres, label='Quick Sort', marker='.')
plt.plot(tamanhos, TempoInsertionSortDecres, label='Insertion Sort', marker='.')
plt.ylabel('Tempo de execução (s)')
plt.xlabel('Tamanho do vetor')
plt.title('Vetor Decrescente')
plt.legend()
plt.show()

plt.plot(tamanhos, TempoQuickSortAlea, label='Quick Sort', marker='.')
plt.plot(tamanhos, TempoInsertionSortAlea, label='Insertion Sort', marker='.')
plt.ylabel('Tempo de execução (s)')
plt.xlabel('Tamanho do vetor')
plt.title('Vetor Aleatório')
plt.legend()
plt.show()


```
